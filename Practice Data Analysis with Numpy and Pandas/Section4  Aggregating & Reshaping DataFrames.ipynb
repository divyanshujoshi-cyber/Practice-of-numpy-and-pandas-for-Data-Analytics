```python
#In this when we used sum everthing is added to each other wheather its string or integer
#so to cope with that we use the specif name of the column and specific functions
premier_league.groupby(['season','HomeTeam'],as_index=False).agg('sum')
```

```python
#so to cope with that we use the specif name of the column and specific functions
(premier_league.groupby(['season','HomeTeam'],as_index=False)
 .agg({'HomeGoals':['sum','mean'],
       'AwayGoals':['sum','mean']}))
```

```python
#the named aggregationn
premier_league.groupby(['season','HomeTeam']).agg(
    home_goal_sum = ('HomeGoals','sum'),
    away_goal_sum= ('AwayGoals','sum'))
```

```python
#transform
pm = premier_league.assign(
avg_teams_goals=premier_league.groupby(['HomeTeam'])['HomeGoals'].transform('mean'),
difference =lambda x: x['HomeGoals']-x['avg_teams_goals'])
```

```python
pm.groupby(['HomeTeam','AwayTeam']).agg({'difference':'mean'}).sort_values('difference')
```
