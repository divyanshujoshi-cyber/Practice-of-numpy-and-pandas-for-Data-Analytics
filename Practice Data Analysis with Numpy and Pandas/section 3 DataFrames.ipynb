```python
#numpy select function for multiple conditions and choices
oil.head()
```

```python
#this is just two conditions so its easy to numpy where but if there are mulyiple condition we have to use where functions mulyiple time
#so to avoid those we use numpy select function
oil['buy']=np.where(oil['price']>100,'too_high','buy')
oil
```

```python
oil['buy'].value_counts()
```

```python
#for conditions more than 3 we use np.slect()
conditions =[
    (oil['price']>100),
    (oil['price']<=100) & (oil['price']>50),
    (oil['price']<=50)
]

choices=['dont buy','buy','strong buy']
oil['buy']=np.select(conditions,choices,default = 'missing')
oil
```
